```python
import pandas as pd
df = pd.read_csv('/content/jaws.csv')
df.head()
```

```python
df.columns
```

```python
negative = [ "angry","anxious", "bored", "depressed", "disgusted", "discouraged", "frightened", "furious", "gloomy", "fatigued"]
positive = ["ease", "calm", "content", "ecstatic", "energetic", "enthusiastic", "excited", "inspired", "relaxed", "satisfied"]
```

```python
columns = df.columns
ratings1 = {"Never": 1, "Rarely": 2, "Sometimes": 3, "Quite often":4, "Extremely often": 5}
ratings2 = {"Never": 5, "Rarely": 4, "Sometimes": 3, "Quite often":2, "Extremely often": 1}


x = 0
y = 0
cleanup_nums = {}
for column in columns:
     for str in negative:
            if str in column:
                print(column)
                cleanup_nums.__setitem__(column, ratings2)
                break;
     for str in positive:
            if str in column:
                print(column)
                cleanup_nums.__setitem__(column, ratings1)
                break;             

        


        
print(cleanup_nums)  
```

```python
df = df.replace(cleanup_nums)
df.head()
```

```python
jaws_columns = columns[6:26]
jaws_columns
```

```python
df['Sum'] = df[jaws_columns].sum(axis=1)
df.head()
```

```python
gender = df.groupby('gender')['Sum'].mean()
gender.plot.bar(x='lab', y='val', rot=0)
```

```python
negative_emo = [ "angry","anxious", "bored", "depressed", "disgusted", "discouraged", "frightened", "furious", "gloomy", "fatigued"]
positive_emo = ["ease", "calm", "content", "ecstatic", "energetic", "enthusiastic", "excited", "inspired", "relaxed", "satisfied"]
df['Positive Sum'] = df[positive_emo].sum(axis=1)
df['Negative Sum'] = df[negative_emo].sum(axis=1)

df.head()
```

```python
salary = df.groupby('monthly salary')['Sum'].mean()
salary.plot.bar(x='lab', y='val', rot=0)
```

```python
status = df.groupby('civil status')['Sum'].mean()
status.plot.bar(x='lab', y='val', rot=0)
```

```python
exp = df.groupby('working experience')['Sum'].mean()
exp.plot.bar(x='lab', y='val', rot=0)
```

```python
work = df.groupby('working arrangement')['Sum'].mean()
work.plot.bar(x='lab', y='val', rot=0)
```
